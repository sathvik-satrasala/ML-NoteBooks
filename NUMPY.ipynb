### Reshaping 

This is performed to reshape or change the shape of an array here the data stays the same only the shape of array changes.      
This is done to do compatible operations for ex:

```python
a2
```

```python
a2.shape
```

```python
a3
```

```python
a3.shape
```

```python
# So when we perform their multiplication we get an error
a2*a3
```

This error arose because their shapes are incompatible i.e they are not following broad casting rules which clearly says:

So therefore in order to perform the operation we need to make their shapes equal or one of them to 1 dimension i.e reshape a2

##### reshape(newsize):
This function reshapes the old array into new array given size
Customises shape 

```python
a2.shape
```

```python
reshape_a2=a2.reshape(2,3,1)
reshape_a2
```

```python
# now we can perform multiplication since a2 dimension is 1 and they are compatible
reshape_a2*a3
```
